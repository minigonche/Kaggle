```python
%matplotlib inline

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

import pandas 

#Loads the training data
titanic = pandas.read_csv("data/train.csv")

#------- Data cleaning ---------
# Training
#--- Age
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())

#--- Sex
titanic.loc[titanic["Sex"] == "male", "Sex"] = 0
titanic.loc[titanic["Sex"] == "female", "Sex"] = 1
titanic["Sex"] = titanic["Sex"].fillna(0)

#--- Embarked
titanic["Embarked"] = titanic["Embarked"].fillna("S")
titanic.loc[titanic["Embarked"] == "S", "Embarked"] = 0
titanic.loc[titanic["Embarked"] == "C", "Embarked"] = 1
titanic.loc[titanic["Embarked"] == "Q", "Embarked"] = 2

# ----- Data enhancement --------

# Generating a familysize column
titanic["FamilySize"] = titanic["SibSp"] + titanic["Parch"]
titanic["NameLength"] = titanic["Name"].apply(lambda x: len(x))
```

```python

# ----- Data Visualization --------

v1 = "Age"
v2 = "Parch"

tit = titanic[["Survived",v1,v2]]

ax = tit[tit["Survived"]==0].plot(kind='scatter', x=v1, y=v2,color='DarkRed', label='No Sobrevivo', marker="x");


tit[tit["Survived"]==1].plot(kind='scatter', x=v1, y=v2,color='DarkGreen', label='Sobrevivio', marker="+", ax=ax);

```
